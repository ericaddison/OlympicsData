```python
# parse all countries for country info
# set up imports here
import urllib2
import re
import json
from bs4 import BeautifulSoup
from bs4 import Tag
import sys
from datetime import datetime
```

```python
# parse one country page for info
def parse_country_page(url):
    newCountry = {}
    
    # get the html
    html = urllib2.urlopen(url).read()

    # soup it up
    soup = BeautifulSoup(html)

    # rows
    labels = soup.find_all("span",class_="label")
    
    for l in labels:
        if l.get_text() != "":
            newID = l.get_text().replace(" ","-").strip().lower()
            newCountry[newID] = l.find_next_sibling("div",class_="txt").get_text().replace("\n"," ").strip()
    
    return newCountry
```

```python
# parse the main page for all Sports
baseUrl = "https://www.olympic.org"

# get the html
html = urllib2.urlopen(baseUrl+"/national-olympic-committees").read()

# soup it up
countrySoup = BeautifulSoup(html)

# initliaze olympics dictionary
countries = {}

# find the link to the sports page, which is the beginning of the sports table
countryDiv = countrySoup.find("div",class_="countries")

links = countryDiv.find_all("a")
for l in links:
    newID = l['href'].split("/")[-1].strip()
    countries[newID] = parse_country_page(baseUrl+l['href'])
    countries[newID]['name'] = l.get_text().strip()
    print "done with " + newID
    
with open('CountryTable.json', 'w') as outfile:
    json.dump(countries, outfile, indent=4, sort_keys=True, separators=(',', ':'))
    outfile.close()
```

```python
# a table to relate three letter country identifier to my country keys 
countryIDtable = {}
for c in countries:
    countryIDtable[countries[c]['flag']] = c
with open('CountryIDTable.json', 'w') as outfile:
    json.dump(countryIDtable, outfile, indent=4, sort_keys=True, separators=(',', ':'))
    outfile.close()
```
