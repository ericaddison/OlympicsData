```python
import urllib2
import re
import json
from bs4 import BeautifulSoup
```

```python
# load all the sports
with open('./SportsTable.json') as data_file:
    sportsTable = json.load(data_file)
```

```python
# Need to parse all events from a sports page
# base url for olympics site
baseUrl = "https://www.olympic.org"

# loop through sports
for sport in sportsTable:
    print "parsing events for " + sport
    try:
        # get the html
        html = urllib2.urlopen(baseUrl+"/"+sport).read()

        # soup it up
        sportSoup = BeautifulSoup(html)

        # find the right section
        sportSection = sportSoup.find("section",class_=re.compile("sport-events.*"))

        # get the links from that section, holding the events
        eventLinks = sportSection.find_all("a")

        # save that info
        sportsTable[sport]["events"] = {};
        for event in eventLinks:
            eventID = event['href'].split("/")[-1].strip()
            sportsTable[sport]["events"][eventID] = event.string
    except:
        print "!!! Problem with sport " + sport

with open('SportsTable.json', 'w') as outfile:
    json.dump(sportsTable, outfile, indent=4, sort_keys=True, separators=(',', ':'))
    outfile.close()
```
