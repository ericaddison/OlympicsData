```python
# keys for the objects
olympicKeys = ["Year", "GameCode", "Sport", "SportCode", "Event", "Athlete", "AtlheteCode" "Result", 
               "Medal", "CountryCode"];


# take in a <tr ...> </tr> row of a table
# and parse it. Return an object holding proper info
def ParseCountryTableRow(tableRow):
    columns = tableRow.find_all("td")
    newRowData = {};
    newRowData["Year"] = columns[0].string
    newRowData["GameCode"] = columns[0].a['href'].split("=")[-1]
    newRowData["Sport"] = columns[1].string
    newRowData["SportCode"] =columns[1].a['href'].split("=")[-1]
    newRowData["Event"] = columns[2].string
    newRowData["Athlete"] = columns[3].string
    newRowData["AthleteCode"] = columns[3].a['href'].split("=")[-1]
    newRowData["Result"] = columns[4].string
    newRowData["Medal"] = columns[5].string
    
    return newRowData;


# parse one country page 
def ParseCountryPage(countryPageUrl):
    
    # get the html for the given page and soupify it
    response = urllib2.urlopen(countryPageUrl);
    html = response.read();
    olympicSoup = BeautifulSoup(html);
    
    # find the header of the stats table and init data structures
    statTableHeaderRow = olympicSoup.find_all("tr",class_='statHead')
    dataRows = [];
    
    # loop through siblings of the stat table header,
    # i.e. the data rows of the table
    # and copy the data and the athlete link
    for row in statTableHeaderRow[0].next_siblings:
        if isinstance(row,Tag):
            parsedRow = ParseCountryTableRow(row)
            parsedRow["CountryCode"] = countryPageUrl.split("=")[-1]
            dataRows.append(parsedRow)
            
    # return the list of dataRows and the set of athleteLinks as a list
    return dataRows
```

```python
# try the function
data = ParseCountryPage('http://www.databaseolympics.com/country/countryyear.htm?g=1&cty=GRE')
print "Found",len(data),"entries"
print data[10]
```

```python

# build a list of all the countries and associated country codes
countryListPage = "http://www.databaseolympics.com/country/countrylist.htm"

# get the html for the given page and soupify it
response = urllib2.urlopen(countryListPage);
html = response.read();
countrySoup = BeautifulSoup(html);
countryCodes = {};
allCodes = countrySoup.find_all("a",href=re.compile("^\/country\/countrypage.*"))
for cnt in allCodes:
    code = cnt['href'].split("=")[-1]
    countryCodes[code] = cnt.string
    
print "Imported",len(countryCodes),"country codes"

with open('CountryCodes.json', 'w') as outfile:
    json.dump(countryCodes, outfile, indent=4, sort_keys=True, separators=(',', ':'))
outfile.close()
```

```python
# build a list of all the countries and associated country codes
athleteListPage = "http://www.databaseolympics.com/players/playerlist.htm?lt="

def ParseAtlhetePageList(url, athleteCodes):
    response = urllib2.urlopen(url);
    html = response.read();
    athleteSoup = BeautifulSoup(html);
    allCodes = athleteSoup.find_all("a",href=re.compile("^\/playerpage.htm.*"))
    for cnt in allCodes:
        code = cnt['href'].split("=")[-1]
        athleteCodes[code] = cnt.string
    

athleteCodes = {};
for c in 'abcdefghijklmnopqrstuvwxyz':
    print "Parsing for " + c
    ParseAtlhetePageList(athleteListPage+c,athleteCodes)
    
print len(athleteCodes)

with open('AthleteCodes.json', 'w') as outfile:
    json.dump(athleteCodes, outfile, indent=4, sort_keys=True, separators=(',', ':'))
outfile.close()
```
